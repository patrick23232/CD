```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
#dados = pd.read_excel('PNAD 2013 A.xlsx')
dados = pd.read_table('pes2013A.txt', sep=" ")
```

```python
n=dados.shape[0]
```

```python
list(dados)
```

```python
dados.V0302 = dados.V0302.astype('category')
dados.V0404 = dados.V0404.astype('category')
dados.V6002 = dados.V6002.astype('category')
dados.V4802 = dados.V4802.astype('category')
dados.UF = dados.UF.astype('category')
dados.V1101 = dados.V1101.astype('category')
```

```python
print('Sexo:')
dados.V0302.cat.categories = ('Masculino', 'Feminino')
ut1 = (dados.V0302.value_counts()/n*100).round(decimals=1)
print(ut1,'\n')
plot = ut1.plot(kind='pie',title='Sexo',autopct='%.1f',figsize=(6, 6))
```

```python
print('Cor e Raça:')
dados.V0404.cat.categories = ('Indígena', 'Branca', 'Preta', 'Amarela', 'Parda', 'Sem declaração')
ut2 = (dados.V0404.value_counts()/n*100).round(decimals=1)
print(ut2,'\n')
plot = ut2.plot(kind='bar',title='Cor e Raça',figsize=(6, 6))
```

```python
print('Rede de Ensino:')
dados.V6002.cat.categories = ('Pública', 'Privada')
ut3 = (dados.V6002.value_counts()/n*100).round(decimals=1)
print(ut3,'\n')
plot = ut3.plot(kind='pie',title='Rede de Ensino',figsize=(6, 6))
```

```python
print('Nivel de Ensino:')
dados.V4802 = dados.V4802.astype("category")
dados.V4802.cat.categories = ("Maternal, jardim de infância etc.",
                              "Classe de alfabetização","Alfabetização de adultos",
                              "1ª a 4ª séries do ensino fundamental regular em 8 anos ou de educação de jovens e adultos"
                             ,"5ª a 8ª séries do ensino fundamental regular em 8 anos ou de educação de jovens e adultos"
                             ,"1ª série do ensino fundamental regular em 9 anos",
                              "2ª a 5ª séries do ensino fundamental regular em 9 anos",
                              "6ª a 9ª séries do ensino fundamental regular em 9 anos"
                             ,"Educação de jovens e adultos não seriado do ensino fundamental",
                              "Ensino médio, inclusive pré-vestibular","Superior, inclusive mestrado ou doutorado")
ut4 = (dados.V4802.value_counts()/n*100).round(decimals=1)
print(ut4,'\n')
plot = ut4.plot(kind='bar',title='Nivel de Ensino',figsize=(6, 6))
```

```python
dados.UF.cat.categories = ('Rondonia','Acre','Amazonas','Roraima','Para','Amapa', 'Tocantins', 'Maranhão', 'Piaui','Ceara', 
                           'Rio Grande Do Norte', 'Paraiba', 'pernambuco','Alagoas', 'Sergipe', 'Bahia', 'Minas Gerais', 
                           'Espirito Santo', 'Rio de Janeiro', 'Sao Paulo', 'Paraná', 'Santa Catarina','Rio Grande do Sul', 
                           'Mato Grosso do Sul', 'Mato Grosso', 'Goias', 'Distrito Federal')
ct=pd.crosstab(dados.UF,dados.V4802,margins=True,rownames=['UF'], colnames=['Nivel de Ensino'])
ct
ut4 = (dados.UF[dados.V4802 == 'Ensino médio, inclusive pré-vestibular'].value_counts()/n*100).round(decimals=1)
plot = ut4.plot(kind='pie',title='População no Ensino Médio',figsize=(6, 6))
```

```python
print('Nivel de Ensino por Sexo:')
ct1=pd.crosstab(dados.V0302,dados.V4802,margins=True,rownames=['Sexo'], colnames=['Nivel de Ensino'])
ct1
ut5 = (dados.V0302[dados.V4802 == 'Ensino médio, inclusive pré-vestibular'].value_counts()/n*100).round(decimals=1)
plot = ut5.plot(kind='pie',title='População no Ensino Médio',figsize=(6, 6))
```

```python
ut8 = (dados.V0302[dados.V4802 == 'Superior, inclusive mestrado ou doutorado'].value_counts()/n*100).round(decimals=1)
plot = ut8.plot(kind='pie',title='População no Ensino Superior',figsize=(6, 6))
```

```python
ct1=pd.crosstab(dados.UF,dados.V4802,margins=True,rownames=['UF'], colnames=['Nivel de Ensino'])
ct1
ct1cut=ct1.iloc[:27,9:11]
ct1cut
```

```python
ct1cut.plot.bar(color=('blue','red'),stacked=False, figsize=(10,10))
```

Desse jeito podemos concluir primeiramente que as mulheres são maioria na educação e que investir em cursos com maioria mulheres pode dar certo. 

A segunda analise foi a relação da população do ensino médio e do ensino superior em relação a cada estado do país, alguns estados demonstraram características interessantes, a primeira é a quantidade de estudantes do ensino médio e do ensino superior, como exemplo dessa primeira característica temos: São Paulo, Minas Gerais, Bahia e Rio Grande do Sul, isso pode ter importância para o Sr. Gold, pois ele sabe onde tem maior número de estudantes, focando a criação da faculdade nesses pontos. 
Outra características é a diferença da população no ensino médio para o ensino superior, como: Pará, Maranhão e Tocantins, isso mostra que muitos estudantes do ensino médio não adentram no ensino superior e um desses fatores pode ser a falta de faculdades e universidades. 
Outra característica interessante é que alguns estados contem a população do ensino superior maior que a do ensino médio, por exemplo Distrito Feeral, Mato Grosso do Sul e Santa Catarina, isso significa que esses estados estão recebendo novos estudante para o seu ensino superior, isso pode ser explicado pela migração de pessoas de um estado para outro a procura de boas universidades, e esses estados com boas universidades estão citadas acima. Assim, o Sr. Gold pode evitar esses mesmos estados pois já tem uma universidades de qualidade, e ele pode focar onde a diferença é ao contrário, onde há mais estudante de ensino médio do que superior, como citado na característica anterior.
